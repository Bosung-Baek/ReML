```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

```python
print("tensorflow version : ", tf.__version__)
print("numpy version : ", np.__version__)
```

```python
x = np.arange(0, 2*np.pi, 0.1)
y = np.cos(x)
```

```python
plt.plot(x, y)
plt.show()
```

```python
model = tf.keras.Sequential([
    #input layer using optimized activation function
    tf.keras.layers.Dense(units=1, input_shape=[1]),
    #hidden layer using optimized activation function
    tf.keras.layers.Dense(units=3, activation=tf.nn.sigmoid),
    #output layer using optimized activation function
    tf.keras.layers.Dense(units=1, activation=tf.keras.activations.linear)

])
```

```python
#compile grad optimizer and MSE loss for cos function
model.compile(optimizer=tf.keras.optimizers.Adam(lr=0.01), loss='mse')
# mini batch size = 21 and epochs = 30
model.fit(x, y, batch_size=21, epochs=1000, verbose = 1)
```

```python
#loss function plot
plt.plot(model.history.history['loss'])
```

```python
#predict y values
y_pred = model.predict(x)
plt.plot(x, y_pred)
plt.show()
```
